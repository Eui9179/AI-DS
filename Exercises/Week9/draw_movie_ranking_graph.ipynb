```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

date_list = pd.date_range('2020-07-01', periods=100, freq='D')

movie_title = []
movie_point = []
movie_date = []

for today in date_list:
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=' \
        + (today.strftime('%Y%m%d'))
    html = urlopen(url)
    
    soup = BeautifulSoup(html.read(), 'html.parser')
    
    titles = soup.find_all('div', {'class':'tit5'})
    points = soup.find_all('td', {'class':'point'})
    
    for t, p in zip(titles, points):
        movie_title.append(t.text.strip(' \n'))
        movie_point.append(float(p.text))
        movie_date.append(today)
```

```python
frame = pd.DataFrame({
    'date': movie_date,
    'title': movie_title,
    'point':movie_point
})
```

```python
expr = "title == ['토이 스토리 3']"
search_movie = frame.query(expr)
search_movie
```
