```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm

date_list = pd.date_range('2020-07-01', periods=100, freq='D')

movie_title = []
movie_point = []
movie_date = []

for today in tqdm(date_list):

    url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=" + (today.strftime('%Y%m%d'))
    html = urlopen(url)

    soup = BeautifulSoup(html.read(), "html.parser")
    titles = soup.find_all('div', {'class': 'tit5'})
    points = soup.find_all('td', {'class': 'point'})

    titles_text = []
    points_text = []
    date = []
                                                                               
    for title, point in zip(titles, points):
        titles_text.append(title.text.strip('\n'))
        points_text.append(float(point.text))
        date.append(today)
        
    movie_title += titles_text
    movie_point += points_text
    movie_date += date

print(movie_title)
```

```python
frame = pd.DataFrame({'date': movie_date, 'title': movie_title, 'point': movie_point})
frame
```

## pd.pivot_table

- 피벗 테이블: 커다란 표의 데이터를 요약하는 통계표
- 요약 방식: 합계, 평균, 기타 통계 등 의미 있는 방식으로 묶음
- 데이터 처리의 한 기법: 유용한 정보에 집중할 수 있도록 하기 위해 통계를 정렬 또는 재정렬(피벗)하는것

```python
import numpy as np

pd.options.display.float_format = '{:.2f}'.format
movie_point = pd.pivot_table(
    frame, 
    values=['point'], 
    index=['title'], 
    aggfunc=np.sum
    # sum 합계
    # mean: 평균
)

movie_point.head(10)
```

```python
movie_point2 = pd.pivot_table(
    frame, 
    values=['point'], 
    index=['title'], 
    aggfunc=np.mean
)

movie_point2_sort = movie_point2.sort_values(by='point', ascending=False)
movie_point2.head(10)
```
