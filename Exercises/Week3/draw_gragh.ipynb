# matplotlib 기본

```python
import matplotlib.pyplot as plt

# plt.rc('font', family='Nanum Gothic')
plt.title('Test') # 제목
plt.plot([1, 2, 3 ,4], [12,45,10,1])
plt.show()
```

# plot parameter

## 1. color
라인 색 지정
## 2. label
선 이름 지정

## 3. ls
라인 스타일(점선, ..)

## 4. marker
점에 해당하는 값들을 마킹

```python
import matplotlib.pyplot as plt

plt.title('Test')
plt.plot([1, 2, 3 ,4], [12,45,10,1], color="red", label="test1", ls="--") # 색(color), 라벨(label), 라인스타일(ls)
plt.plot([10,43, 5 ,7], [12,45,10,1], color="blue", label="test2")
plt.legend(loc=5) # 라벨 위치
plt.show()
```
