```python
import csv

f = open('incheon_weather.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

result = []

for row in data:
    high = row[-1]
    result.append(high)

f.close()
```

```python
import matplotlib.pyplot as plt

plt.plot(result, color='red')
plt.show()
```

```python
import csv

f = open('incheon_weather.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

result = []

for row in data:
    date = row[0].split('-')
    year = int(date[0])
    month = date[1]
    day = date[2]
    
    if year >= 1997 and month == '05' and day == '30':
        if row[-1] != '':
            result.append(float(row[-1]))
    
plt.plot(result)
plt.show()
    
```
