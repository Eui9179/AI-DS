plt.plot(years, high_list, label='최고기온', color='#FF796C')

## 2. x축 이름, y축 이름 지정하기

- x축: `plt.xlabel('X-Label')`, String 타입으로 값을 받는다. x축 밑에 표시
- y축: `plt.xlabel('Y-Label')`, String 타입으로 값을 받는다. y축 옆에 표시


- 실제 코드

```python
plt.xlabel('년도')
plt.ylabel('최저기온, 최고기온')
```


## 3. 다른 한글 폰트 사용해보기

- 나눔 고딕

```python
plt.rc('font', family='NanumGothic')
```

````python
import csv
import matplotlib.pyplot as plt

f = open('incheon_weather.csv', 'r', encoding='cp949') # 한글 인코딩(cp949)

data = csv.reader(f)
header = next(data)

low_list = [] # 최저 기온 리스트
high_list = [] # 최고 기온 리스트
years = [] # 년도(x축)
    
for row in data:
    date = row[0].split('-') # 날짜
    year = int(date[0])
    month = date[1] 
    day = date[2]

    if year >= 1997 and month == '05' and day == '30': 
        low = row[-2] # 최저 기온
        high = row[-1] # 최고 기온
        
        if low != '' and high != '':
            low_list.append(float(low))
            high_list.append(float(high))
            years.append(year)

plt.rc('font', family='NanumGothic') # 한글 폰트 적용
plt.title('매년 생일 기온 데이터') # 그래프 제목
plt.xlabel('년도') # x축 라벨
plt.ylabel('최저기온, 최고기온') # y축 라벨
plt.plot(years, high_list, label='최고기온', color='#FF796C') # x축: year, y축: high_list, color: 빨간색
plt.plot(years, low_list, label='최저기온', color='#7BC8F6') # x축: year, y축: low_list, color: 하늘색

plt.legend() # 각 그래프 라벨 표시

plt.show()
````
