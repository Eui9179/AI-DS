# Scatter 그래프로 표현하기

```python
import matplotlib.pyplot as plt
import csv
```

```python
plt.style.use('ggplot')
x = [1, 2, 3, 4]
y = [10, 30, 20, 40]
size = [100, 200, 250, 300]

plt.scatter(x, y, s=size, c=range(4), cmap='jet')
plt.show()
```

# scatter parameter

- s: 각 점의 크기
- c: 각 점의 색깔
- cmap: 색 팔레트
- alpha: 색 투명도

```python
import random

x, y = [], []
size = []

for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))

plt.scatter(x, y, s=size, c=size, cmap='jet')
plt.show()
```

```python
f = open('incheon_population.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

male = []
female = []

for row in data:
    if '송도1동' in row[0]:
        for i in range(3, 104):
            male.append(int(row[i]))
            female.append(int(row[i+103]))
        break

f.close()

plt.rc('font', family='NanumGothic')
plt.title('연령대별 성별 인구 그래프')
plt.scatter(male, female, c=range(101), alpha=0.5, cmap='jet')

plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')

plt.show()
```

# 추세선 추가하기

```python
f = open('incheon_population.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

male = []
female = []

for row in data:
    if '송도1동' in row[0]:
        for i in range(3, 104):
            male.append(int(row[i]))
            female.append(int(row[i+103]))
        break

f.close()

plt.rc('font', family='NanumGothic')
plt.title('연령대별 성별 인구 그래프')
plt.scatter(male, female, c=range(101), alpha=0.5, cmap='jet')

line = range(max(male)) # 남성 최대값 기준
plt.plot(line, line, color='g')

plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')

plt.show()
```
