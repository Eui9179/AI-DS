## 1. 꺾은선 그래프

````python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

header = next(data)

male = []
female = []

for row in data:
    if '송도1동' in row[0]:
        for m in row[3:104]:
            male.append(float(m))
        
        for fem in row[106: 207]:
            female.append(float(fem))
            
plt.rc('font', family='Malgun Gothic')
plt.title('인천 인구 통계')
plt.plot(male, label='male')

plt.legend()
plt.show()

f.close()
````

````python
import matplotlib.pyplot as plt
import csv

f = open('incheon_population.csv', 'r')

data = csv.reader(f)
header = next(data)

result = []

for row in data:
    if '송도1동' in row[0]:
        for m, fm in zip(row[3: 104], row[106: 207]): 
        # zip: 동일한 개수로 이루어진 자료형을 묶어 주는 역할
            result.append(int(m) + int(fm))
        break
        
f.close()
    
plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
plt.title('송도1동 인구')
plt.xlabel('연령')
plt.ylabel('인구수')
plt.plot(result)

plt.show()
````

## 2. 막대 그래프/항아리 그래프

### 막대 그래프

- `bar(막대를 표시할 위치, 막대의 높이)` 함수

````python
import matplotlib.pyplot as plt

plt.bar(range(5), [10, 30, 20, 50, 40])
plt.show()
````

````python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv')
data = csv.reader(f)
header = next(data)

result = []

for row in data:
    if '송도1동' in row[0]:
        for m, fm in zip(row[3: 104], row[106: 207]):
            result.append(int(m) + int(fm))
        break
        
f.close()

plt.rc('font', family='Malgun Gothic')
plt.title('인구 통계')

plt.bar(range(101), result)
plt.xlabel('연령')
plt.ylabel('인구수')

plt.show()
````

수평 막대그래프(`barh()` 함수)를 활용해 우리 동네 인구 구조 시각화하기

```python
plt.barh(range(101), result)
```

````python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv')
data = csv.reader(f)
header = next(data)

result = []

for row in data:
    if '송도1동' in row[0]:
        for m, fm in zip(row[3: 104], row[106: 207]):
            result.append(int(m) + int(fm))
        break

plt.rc('font', family='Malgun Gothic')
plt.title('인구 통계')

plt.barh(range(101), result)
plt.xlabel('연구수')
plt.ylabel('연령')
plt.show()
````

````python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)
header = next(data)

male = []
female = []

for row in data:
    if '송도1동' in row[0]:
        for m, fm in zip(row[3:104], row[106:207]):
            male.append(-int(m))
            female.append(int(fm))

plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호
plt.xlabel('인구수')
plt.ylabel('연령')
plt.barh(range(101), male, color='b', label='남자')
plt.barh(range(101), female, color='r', label='여자')
plt.legend()
plt.show()
````

## 3. 파이차트

### pie() 함수

- 파이 차트(pie chart)는 '원 그래프'

````python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.axis('equal') # 동그란 원
plt.pie([10, 20, 30, 40])
plt.show()
````

````python
import matplotlib.pyplot as plt

size = [100, 200, 300, 400]
labels = ['p1', 'p2', 'p3', 'p4']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink'] # 색상
exp = (0, 0, 0.1, 0) # 돌출 효과

plt.pie(
    size, 
    labels=labels, # 레이블
    autopct='%.1f%%', # autopct = 퍼센트
    colors=color, # 색깔
    explode=exp, # 돌출
)
plt.legend()
plt.show()
````

````python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

header = next(data)

m_cnt = 0
fm_cnt = 0

for row in data:
    if '송도1동' in row[0]:
        for m, fm in zip(row[3:104], row[106:207]):
            m_cnt += int(m)
            fm_cnt += int(fm)
        break

size = [m_cnt, fm_cnt]
label = ['남성', '여성']

plt.rc('font', family='Malgun Gothic')
plt.axis('equal')
plt.pie(
    size, 
    labels=label, 
    autopct='%.1f%%', 
    explode=(0.1, 0), # 돌출
    startangle=90, # 각도
)
plt.show()
````

### 버블 차트로 표현하기

- `scatter()` 함수를 이용하여 버블 차트로 그릴 수 있음
- 기존 코드에 size를 의미하는 s 속성을 추가하고 원하는 크기 입력


````python
import matplotlib.pyplot as plt

plt.scatter(
    [3,1,2,4], 
    [10, 20, 30, 40], 
    s=[100, 200, 450, 230],
    c=range(4), # colorb
    cmap='jet' # color map
)
# plt.colorbar()
plt.show()
````

````python
import random
import matplotlib.pyplot as plt

x, y = [], []
size = []

for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))

plt.scatter(x, y, s=size, c=size, cmap='jet', alpha=0.7)
# plt.colorbar()
plt.show()
````

````python
import csv, math
import matplotlib.pyplot as plt

f = open('incheon_population.csv')
data = csv.reader(f)

years = []
male = []
female = []
size = []

for row in data:
    if '송도1동' in row[0]:
        for m, fm in zip(row[3:104], row[106:207]):
            male.append(int(m))
            female.append(int(fm))
            size.append(math.sqrt(int(m)+int(fm)))
        break

plt.scatter(male, female, c=range(101), alpha=0.7, cmap='jet', s=size)
line = range(max(male))
plt.plot(line, line, color='k')
# plt.colorbar()
plt.show()
````
