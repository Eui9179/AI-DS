```python
import csv
import matplotlib.pyplot as plt

f = open('mypopulation.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

male = []
female = []
result = []

for row in data:
    if '마전동' in row[0]:
        for m, fm in zip(row[3:104], row[106:206]):
            male.append(int(m))
            female.append(int(fm))
            result.append(int(m) + int(fm))

f.close()
            
plt.rc('font', family='NanumGothic')            
plt.title('마전동 연령대별 인구수')
plt.style.use('ggplot')
plt.plot(male, color='b', label='남성')
plt.plot(female, color='r', label='여성')
plt.plot(result, color='black', label='전체')

plt.xlabel('연령')
plt.ylabel('인구수')
plt.legend()
f.close()
plt.show()

```
