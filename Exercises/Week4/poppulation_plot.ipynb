```python
import csv
import matplotlib.pyplot as plt
f = open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)
total = next(data)

male = []
female = []

result = []
for row in data:
    if '송도1동' in row[0]:
        for m, fm in zip(row[3:104], row[106:206]):
            male.append(int(m))
            female.append(int(fm))
            result.append(int(m) + int(fm))

plt.rc('font', family='NanumGothic')            
plt.title('송도 1동 연령대별 인구수')
plt.style.use('ggplot')
plt.xlabel('연령')
plt.ylabel('인구수')
plt.plot(male, color='b')
plt.plot(female, color='r')
plt.plot(result, color='c')
    
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('알고 싶은 동네 입력: ')

male = []
female = []

result = []
for row in data:
    if name in row[0]:
        for m, fm in zip(row[3:104], row[106:206]):
            male.append(int(m))
            female.append(int(fm))
            result.append(int(m) + int(fm))

plt.rc('font', family='NanumGothic')            
plt.title(f'{name} 연령대별 인구수')
plt.style.use('ggplot')
plt.xlabel('연령')
plt.ylabel('인구수')

plt.plot(male, color='b')
plt.plot(female, color='r')
plt.plot(result, color='c')
    
plt.show()
```
