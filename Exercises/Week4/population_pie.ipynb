# Pie 그래프 그리기

```python
import matplotlib.pyplot as plt
import csv
```

```python
plt.rc('font', family='NanumGothic')
plt.axis('equal') # 동그란 원으로 그리기

size = [10, 20, 30, 40]
label = ['레이블1', '레이블2', '레이블3', '레이블3']

exp = [0, 0.1, 0, 0]

plt.pie(size, labels=label, autopct='%.2f%%', explode=exp) # 
plt.legend()
plt.show()
```

## plt.pie parameter
- explode: 그래프 강조하기 (default=0)
- autopct: 그래프 퍼센트 자동 계산 (%.{num}f%% 소수 자리수)

# 송도 1동 남녀 인구수

```python
f = open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

male = 0
female = 0

for row in data:
    if '송도1동' in row[0]:
        male = int(row[1].replace(',', ''))
        female = int(row[104].replace(',', ''))
        break

plt.rc('font', family='NanumGothic')

result = [male, female]
label = ['남성', '여성']
plt.pie(result, labels = label, autopct='%.1f%%')

plt.show()
```
