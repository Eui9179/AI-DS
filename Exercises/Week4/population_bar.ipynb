# Bar 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt
```

```python
plt.bar(range(6), [1, 2, 3, 5, 6, 7])
plt.show()
```

```python
f = open('incheon_population.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

result = []

for row in data:
    if '송도1동' in row[0]:
        for m, fm in zip(row[3:104], row[106:207]):
            result.append(int(m) + int(fm))
```

```python
plt.rc('font', family="NanumGothic")
plt.title('인구 통계')

plt.bar(range(101), result)
plt.xlabel('연령')
plt.ylabel('인구수')

plt.show()
 
```

# 항아리 그래프 그리기

```python
f = open('incheon_population.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data)

m = []
fm = []

for row in data:
    if '송도1동' in row[0]:
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:207]:
            fm.append(int(i))
            
f.close()

plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), m, label='남성')
plt.barh(range(101), fm, label='여성')
plt.xlabel('연령')
plt.ylabel('인구수')

plt.legend()
plt.show()
```
